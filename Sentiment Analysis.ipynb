### Load the files with positive, negative, stop words

```python
# --------To read positive words from a text file and put it in a list--------
positive = [word.strip() for word in open('positive.txt','r') ]
print(positive)
```

```python
# --------To read negative words from a text file and put it in a list--------
negative = [word.strip() for word in open('negative-words.txt','r')]
print(negative)
```

```python
# --------To read stop words from a text file and put it in a list--------
stop = [word.strip() for word in open('stopwords.txt','r') ]
print(stop)
```

### Dictionary of positive,negative words

Creating a dictionary of positive, negative words and assigning value 1 to positive words, -1 to negative words

```python
# --------To make a single dictionary of positive and negative words.
d = {}
# Each positive word is a key in the dictionary and its value is assigned to be +1 using above for loop
for word in positive:
    d[word] = 1
# Each negative word is a key in the dictionary and its value is assigned to be -1 using above for loop
for word in negative:
    d[word]=-1
# Prints dictionary of words as keys and values as +1 for positive word and -1 for negative word
print(d)
```

### Extracting the text file containing tweets 

```python
with open('Trump.txt', 'r') as my_file:
    data = my_file.read()
print(data)
```

### Saving the data as a list

```python
# Prints list of tweets by splitting each sentence into set of words.
tweets = data.split(" ")
print(tweets)
```

### Cleaning the data, Calculating the count of words and ratio:

```python
import re
def cal_sentiment_analysis(tweet_data):
# Create and initialize a variable to count number of positive, negative, stop and other words
    positive_count=0
    negative_count = 0
    stop_count = 0
    other_count = 0
    Total_word_count = 0
    for word in tweet_data:
        w1 = word.lower().strip("\" !#@,./\_][() >:';&$?") # remove punctuations attached to each word
        text = re.sub(r'^https?:\/\/.*[\r\n]*', '', w1, flags=re.MULTILINE)   # removing URL
        Total_word_count+=1
        if(text not in stop):                                           # If word not an stop word
            if (text in d.keys()):                                      # If clean word is present in the dictionary
                if (d[text] == 1):                                      # if it is a positive word
                    positive_count += 1                                 # increase the count of positive word               
                elif (d[text] == -1):                                   # if it is a negative word
                    negative_count += 1                                 # increase the count of negative word
            else:                                                       # else
                other_count+=1                                          # increase the count of other word
        if(text in stop):
            stop_count+=1
    
    # Print number of words
    print('Total word count: ' + str(Total_word_count))              
    print('Total number of positive words: ' +str(positive_count))
    print('Total Number of negative words: ' +str(negative_count))
    print('Total number of stop words: ' +str(stop_count))
    print('Total number of other words: ' +str(other_count))
    
    # Calculating Ratios
    print('\nThe ratio of postive word compared to the whole word count is: '+str(round(positive_count/Total_word_count,3)))
    print('The ratio of negative word compared to the whole word count is: '+str(round(negative_count/Total_word_count,3)))
    print('The ratio of stop word compared to the whole word count is: '+str(round(stop_count/Total_word_count,3)))
    print('The ratio of other word compared to the whole word count is: '+str(round(other_count/Total_word_count,3)))
    
    # sum of postive and negative words
    sum_of_positive_negative = (positive_count+negative_count)
    print('The sum of all postive or negative words observed:' +str(sum_of_positive_negative))
    
    
    # Overall sentiment
    if(positive_count>negative_count):
        print('\033[1m'+'\nOverall sentiment of the tweets is a positive number:'+str(positive_count-negative_count))
        print('As the sentiment is positive, percentage of positive words: '+ str(positive_count*100/sum_of_positive_negative)+'%')
        if((positive_count*100/sum_of_positive_negative) > 50):
            print('\033[1m'+'\nWe conclude that the overall sentiment is Strongly positive')
        else:
            print('\nWe conclude that the overall sentiment is Weakly Positive')
    else:
        print('\033[1m'+'\nOverall sentiment of the tweets is a negative number:'+str(negative_count - positive_count))
        print('As the sentiment is negative, percentage of positive words: '+ str(negative_count*100/sum_of_positive_negative)+'%')
        if((negative_count*100/sum_of_positive_negative) > 50):
            print('\033[1m'+'\nWe conclude that the overall sentiment is Strongly Negative')
        else:
            print('\nWe conclude that the overall sentiment is Weakly Negative')
    
```

### Calling function cal_sentiment_analysis

```python
cal_sentiment_analysis(tweets)
```

### Is the general sentiment positive or negative?


The total number of positive words is: 683.

The total number of negative words is: 567.

The overall sentiment of the tweets in numeric value is: 683-567 = 116.

As the overall sentiment of the tweets is a positive number, we can say that the general sentiment is positive.

### Is the sentiment strongly or weakly positive?

Lets define the sentiment level into 4 parts:

0 % - 50: Weak

50% - 100%: Strong

The total number of words is 15614.

As the sentiment is positive, the percentage of positive words: (683/1250)*100 = 54.64%.

As 4.37% lies in the 1st part of our sentiment level, we can conclude that the overall sentiment is weak.
